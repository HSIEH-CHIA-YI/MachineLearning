```python
from google.colab import userdata #在google.colab放入userdata

ngrok_authtoken = userdata.get('NGROK_AUTHTOKEN') #從Colab的userdata裡面讀取一個叫NGROK_AUTHTOKEN的值
port=5051 #設定本機的埠號為5051
```

```python
import os
from pyngrok import ngrok #在Python的內建的模組套件pyngrok匯入ngrok這個工具
```

```python
ngrok.kill() #關閉所有正在運行的 ngrok 連線
```

```python
ngrok.set_auth_token(ngrok_authtoken) #從Colab讀到的token設定給ngrok
tunnel = ngrok.connect(port, name="linebot_tunnel") #幫這個連線取名字
print(tunnel.public_url) #印出你的公開網址
```

```python
from flask import Flask   #使用Flask建立一個網站應用

app = Flask(__name__)

@app.route('/') #當有人訪問「首頁 /」時，就會執行 home() 這個函式
def home():
    return "<h1>明新科技大學 謝佳儀</h1>"

if __name__ == '__main__': #如果這個__main__是被直接執行，就啟動網站
    app.run(port=port)  #開始監聽5051 port，讓別人可以連進來
```
